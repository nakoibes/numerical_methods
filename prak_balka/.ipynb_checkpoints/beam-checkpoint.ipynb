```python
var("z w0 th0 M0 Q0 q1 q2 R1 R2 R3 E_sup k J E")


J = 10e-6;
E = 2e11;
q1 = 100.0;
q2 = 800.0;
k = 20000.0;
M1 = -1000.0;
P = -700;

Px = 2
R1x = 4
R2x = 11
R3x = 19
q1x = 8
q2x = 20
qlen = q2x-q1x
Mx = 21
kx = 23

with open("dots.txt", "w") as f:
    f.write(str(Px))
    f.write("\n")
    f.write(str(R1x))
    f.write("\n")
    f.write(str(R2x))
    f.write("\n")
    f.write(str(R3x))
    f.write("\n")
    f.write(str(q1x))
    f.write("\n")
    f.write(str(q2x))
    f.write("\n")
    f.write(str(qlen))
    f.write("\n")
    f.write(str(Mx))
    f.write("\n")
    f.write(str(kx))

function_zero(z) = 0

w1(z) = w0 + th0 * z +(M0 * (z^2/2) + Q0 * (z^3/6)) / (E * J)
w2(z) = heaviside(z-Mx) * M1 * ((z-Mx)^2/2) / (E * J)
w3(z) = heaviside(z-kx) * E_sup * ((z-kx)^3/6) / (E * J)
w4(z) = heaviside(z-R1x) * R1 * ((z-R1x)^3/6) / (E * J)
w5(z) = heaviside(z-R2x) * R2 * ((z-R2x)^3/6) / (E * J)
w6(z) = heaviside(z-R3x) * R3 * ((z-R3x)^3/6) / (E * J)
w7(z) = heaviside(z-q1x) * q1 * ((z-q1x)^4/24) / (E * J)
w8(z) = heaviside(z-q1x) * ((q2-q1)/qlen) * ((z-q1x)^5/120) / (E * J)
w9(z) = -heaviside(z-q2x) * q2 * ((z-q2x)^4/24) / (E * J)
w10(z) = -heaviside(z-q2x) * ((q2-q1)/qlen) * ((z-q2x)^5/120) / (E * J)
w11(z) = heaviside(z-Px) * P * ((z-Px)^3/6) / (E * J)


w(z) = w1(z) + w2(z) + w3(z)+ w4(z)+w5(z) + w6(z) +w7(z) + w8(z)+w9(z) + w10(z)+ w11(z)


th = diff(w, z).substitute_function(dirac_delta, function_zero)
M = diff(th, z).substitute_function(dirac_delta, function_zero)* (E*J)
Q = diff(M, z).substitute_function(dirac_delta, function_zero)


decision = solve([w(z=0) == 0,
                  th(z=0) == 0,
                  w(z=kx) * k == -E_sup,
                  w(R1x) ==0,
                  w(R2x) ==0,
                  w(R3x)==0,
                  M(z=25) == 0,
                  Q(z=25) == 0,
                 ], 
                  w0, th0, M0,R1,R2,R3, Q0, E_sup, solution_dict=True)

print(decision)
w0 = decision[0][w0]
th0 = decision[0][th0]
M0 = decision[0][M0]
Q0 = decision[0][Q0]
R1 = decision[0][R1]
R2 = decision[0][R2]
R3 = decision[0][R3]
E_sup = decision[0][E_sup]


w1(z) = w0 + th0 * z +(M0 * (z^2/2) + Q0 * (z^3/6)) / (E * J)
w2(z) = heaviside(z-Mx) * M1 * ((z-Mx)^2/2) / (E * J)
w3(z) = heaviside(z-kx) * E_sup * ((z-kx)^3/6) / (E * J)
w4(z) = heaviside(z-R1x) * R1 * ((z-R1x)^3/6) / (E * J)
w5(z) = heaviside(z-R2x) * R2 * ((z-R2x)^3/6) / (E * J)
w6(z) = heaviside(z-R3x) * R3 * ((z-R3x)^3/6) / (E * J)
w7(z) = heaviside(z-q1x) * q1 * ((z-q1x)^4/24) / (E * J)
w8(z) = heaviside(z-q1x) * ((q2-q1)/qlen) * ((z-q1x)^5/120) / (E * J)
w9(z) = -heaviside(z-q2x) * q2 * ((z-q2x)^4/24) / (E * J)
w10(z) = -heaviside(z-q2x) * ((q2-q1)/qlen) * ((z-q2x)^5/120) / (E * J)
w11(z) = heaviside(z-Px) * P * ((z-Px)^3/6) / (E * J)


w(z) = w1(z) + w2(z) + w3(z) +w4(z)+w5(z) + w6(z) +w7(z) + w8(z)+w9(z) + w10(z)+ w11(z)


th(z) = diff(w, z).substitute_function(dirac_delta, function_zero)
M(z) = diff(th, z).substitute_function(dirac_delta, function_zero) * (E*J)
Q(z) = diff(M, z).substitute_function(dirac_delta, function_zero)
with open('sage_output.txt', 'w') as out:
    x = 0.0
    for i in range(1001):
        x += (0.025)
        print(x, w(x), th(x), M(x), Q(x), sep = ',', file = out)
```
